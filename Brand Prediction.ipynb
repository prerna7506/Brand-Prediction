```python
def get_matching_brands(reference_brand, brands_to_compare, threshold=65):
    matching_brands = []

    for brand in brands_to_compare:
        match_percentage = fuzz.ratio(reference_brand, brand)
        if match_percentage > threshold:
            matching_brands.append(brand)

    return matching_brands
```

```python
def get_matching_brands_for_multiple_references(reference_brand_dict, brands_to_compare, threshold=70):
    result_dict = {}


    for reference_brand in reference_brand_dict:
        matching_brands = get_matching_brands(reference_brand, brands_to_compare, threshold)
        result_dict[reference_brand] = sorted(set(matching_brands))

    return result_dict
```
